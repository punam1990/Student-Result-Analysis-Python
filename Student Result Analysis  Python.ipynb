```python
# IMPORTING LIBRARIES


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Expanded_data_with_more_features.csv")
print(df)
```

```python
# DROP UNNAMED COLUMN

df= df.drop("Unnamed: 0", axis =1)
print(df.head())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May","5-10")
df.head()
```

```python
#DATA VISUALIZATION 
#GENDER DISDTRIBUTION


plt.figure(figsize= (5,5))
ax = sns.countplot(data= df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#FROM THE ABOVE CHART WE HAVE ANALYSED THAT:
#THE NUMBER OF FEMALE IS MORE THAN THE NUMBER OF MALE
```

```python
#PARENT'S EDUCATION IMPACT ON STUDENT SCORES

gb= df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#FROM THE ABOVE CHART WE HAVE CONCLUDED THAT THE EDUCATION OF THE PARENTS 
#HAVE GOOD IMPACT ON STUDENTS SCORE
```
